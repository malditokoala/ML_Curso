```python
fig, ax = plt.subplots()

hb =ax.hexbin(df["Latitud"], df["Longitud"],C=df["Luminosidad"], gridsize=50, cmap="Greens")
ax.set_facecolor('#000000')
plt.colorbar(hb, label='Nivel de sombras')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.title('Luminosidad de en Villa Luz')
;
```
