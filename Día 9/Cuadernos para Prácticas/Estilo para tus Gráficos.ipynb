```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 35, 26, 33]
    })
df
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

```python
plt.style.use('ggplot')
fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

```python
df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

plt.bar(df["frutas"], df["cantidades"])
plt.title('Distribución de frutas en el almacén')

plt.show()
```

```python
# Crear el DataFrame con los datos
df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

plt.pie(df["cantidades"], labels=df["frutas"], autopct="%1.1f%%", explode=(0.1, 0.1, 0.1, 0.1))
plt.title('Distribución de frutas en el almacén')
plt.style.use('dark_background')
plt.show()
```
