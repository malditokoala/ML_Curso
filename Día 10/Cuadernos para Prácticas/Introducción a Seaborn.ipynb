```python
# Importar seaborn
import seaborn as sns

# Aplicar el tema por defecto
sns.set_theme()

# Cargar un dataset de ejemplo
propinas = sns.load_dataset("tips")

# Crear una visualizacin
sns.relplot(
    data=propinas,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker",
    style="smoker",
    size="size",
);
```
